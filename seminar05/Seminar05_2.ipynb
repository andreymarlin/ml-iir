```python
df = pd.DataFrame(list(documents.items()), columns=['doc_id', 'doc_content'])
```

```python
print(df.head())
```

```python
# Загрузка модели (tiny-версия)
model = SentenceTransformer('cointegrated/rubert-tiny2').to('cuda')
```

```python
# Создаем эмбеддинги для всех документов
embeddings = model.encode(df['doc_content'].tolist(), convert_to_numpy=True)
df['embeddings'] = list(embeddings)
```

```python
df.head()
```

```python
def find_top_n(query, dataframe, top_n=5):
    query_embedding = model.encode([query])

    doc_embeddings = np.stack(dataframe['embeddings'].values)

    scores = cosine_similarity(query_embedding, doc_embeddings)[0]

    results = dataframe.copy()
    results['score'] = scores
    return results.sort_values(by='score', ascending=False).head(top_n)
```

```python
query = "Аналитика и отчетность"
top_5 = find_top_n(query, df, top_n=5)
```

```python
print(f"Запрос: {query}")
print(top_5[['doc_id', 'doc_content', 'score']])
```
